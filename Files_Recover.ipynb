```python
!pip install filetype
import os
import filetype
import shutil
from google.colab import drive

# 1. Mount Drive
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')

# 2. Set Your Paths
# Ensure this matches the EXACT name of the folder you just uploaded
source_folder = '/content/drive/MyDrive/Siddhi_Recovered_files'
output_folder = '/content/drive/MyDrive/Siddhi_Recovered_Final'

if not os.path.exists(output_folder):
    os.makedirs(output_folder)

print(f"ðŸš€ Processing files in: {source_folder}")

# 3. The Recovery Loop
found_count = 0
for root, dirs, files in os.walk(source_folder):
    for filename in files:
        if filename.upper().endswith(".CHK") or filename == "Siddhi":
            found_count += 1
            file_path = os.path.join(root, filename)

            # Detect file type via Magic Bytes
            kind = filetype.guess(file_path)

            if kind:
                ext = f".{kind.extension}"
            else:
                # Fallback for your Python/R code (Vigilance Alert/SpeedSync projects)
                try:
                    with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                        content = f.read(1000)
                        if "import " in content or "def " in content:
                            ext = ".py"
                        elif "library(" in content or "<-" in content:
                            ext = ".R"
                        elif "," in content:
                            ext = ".csv"
                        else:
                            ext = ".txt"
                except:
                    ext = ".dat"

            # Save the fixed version
            new_name = filename.split('.')[0] + ext
            shutil.copy(file_path, os.path.join(output_folder, new_name))
            print(f"âœ… Fixed: {filename} -> {new_name}")

if found_count == 0:
    print("âŒ No files found. Please check if the folder name on Drive matches 'source_folder'.")
else:
    print(f"\nðŸŽ‰ SUCCESS! {found_count} files are ready in 'Siddhi_Recovered_Final'")
```
